# Read

```python
dados_conexao = ("Driver={SQLite3 ODBC Driver};"
                 "Server=localhost;"
                 "Database=db/chinook.db;")

conexao = pyodbc.connect(dados_conexao)
cursor = conexao.cursor()
```

```python
cursor.execute("SELECT * FROM customers")

valores = cursor.fetchall()
descricao = cursor.description
cursor.close()
print(type(valores))
```

```python
import pandas as pd 
colunas = [desc[0] for desc in descricao]
tabela_clientes = pd.DataFrame.from_records(valores, columns= colunas)
display(tabela_clientes.head())
```

```python
conexao.close()
```

```python
dados_conexao = ("Driver={SQLite3 ODBC Driver};"
                 "Server=localhost;"
                 "Database=db/salarios.sqlite;")

conexao = pyodbc.connect(dados_conexao)
cursor = conexao.cursor()
```

```python
tabela_cliente = pd.read_sql("SELECT * FROM Salaries", conexao)
display(tabela_cliente.head())
```

```python
conexao.close()
```

```python
tabela_cliente_editada = tabela_cliente.loc[tabela_cliente['Agency'] == 'San Francisco']
display(tabela_cliente_editada)
```

```python
tabela_qnt = tabela_cliente.groupby('Year').count()
display(tabela_qnt)
```

```python
tabela_qnt = tabela_cliente.groupby('Year').sum()
tabela_qnt = tabela_qnt[["TotalPay", "TotalPayBenefits"]]
display(round(tabela_qnt,2))
```
